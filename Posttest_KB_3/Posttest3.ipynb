```python
df = pd.read_csv("footballData.csv")
df.info()
```

```python
x = list(df["height_cm"])
y = list(df["age"])

plt.figure(figsize = (5,5))
plt.bar(x, y, label = "Player", color = "red")
plt.title("average height")
plt.legend()
plt.show()
```

Perbandingan Usia dengan tinggi dengan bar (maaf datasetnya kebanyakan bang)

```python
genres = df.dropna(subset=['age']).sort_values(by=['age'])['age'].unique()

total = df.groupby(['age'])['age'].count()

patches, texts, pcts = plt.pie(total, labels=genres, autopct='%.1f%%', wedgeprops={'linewidth': 4.0, 'edgecolor': 'red'},shadow = True, textprops={'size': 'x-large'})

plt.title('Age Comparison')
plt.setp(pcts, color='black', fontweight='bold')
plt.show()
```

Presentase Jumlah Pemain dengan usia tertentu menggunakan pie

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15,15))
sns.scatterplot(data=df,x='age',y='height_cm',hue='weight_kg')
plt.show()
```

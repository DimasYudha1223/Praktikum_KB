```python
# Menampilkan Info Dari Dataset
import pandas as pd
import numpy as np

from prettytable import PrettyTable

df = pd.read_csv("footballData.csv")

df.info()
```

Perintah dari point 1 dan 2 yaitu menampilkan informasi pada dataset yaitu atribut, jumlah atribut dan jumlah attribut yang memiliki nilai null.

```python
dataset_inf = {
    "Jumlah Record Dataset"                 : df.shape[0],
    "Nama Attribute Dataset"               : list(df.columns.values),
    "Jumlah Attribute Tipe Objek Dataset" : list(df.select_dtypes(["object"]).columns),
    "Jumlah Attribute Tipe Angka Dataset"  : list(df.select_dtypes(["int64", "float64"]).columns),
    "Jumlah Attribute Null Dataset"        : list(df.loc[:, df.isnull().any()].columns),
}

keys = list(dataset_inf.keys())

dataset_infTable = PrettyTable()
dataset_infTable.field_names = ["Info", "Atribute", "Jumlah"]

dataset_infTable.add_row([keys[0], "-", dataset_inf[keys[0]]])

for i in range(1, 5):
    dataset_infTable.add_row([keys[i], ", \n".join(dataset_inf[keys[i]]), len(dataset_inf[keys[i]])])
    
print(dataset_infTable)
```
